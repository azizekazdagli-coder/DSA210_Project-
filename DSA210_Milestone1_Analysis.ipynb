## 1. Introduction
This project analyzes the relationship between country-level economic indicators (GDP per capita) and international air traffic patterns across European countries from 2016 to 2024.

### Research Question
To what extent does a country's economic strength (GDP per capita) explain the differences in air traffic volume?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

# Setting aesthetic parameters for plots
sns.set_theme(style="whitegrid")

# Load the merged dataset
# Make sure 'master_dataset_EU_only.csv' is in your directory
try:
    df = pd.read_csv('master_dataset_EU_only.csv')
    print(f"Successfully loaded dataset with {df.shape[0]} rows.")
except FileNotFoundError:
    print("Error: 'master_dataset_EU_only.csv' not found. Please ensure the file is in the same folder.")
```

## 2. Exploratory Data Analysis (EDA)
We start by examining the correlation between traffic variables and economic metrics.

```python
plt.figure(figsize=(10, 8))
corr_cols = ['FLT_TOT_1', 'GDP_pc', 'Population', 'Inbound_Per_Capita', 'Outbound_Per_Capita']
corr_matrix = df[corr_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Analysis: EU Airport Traffic & Economic Indicators')
plt.show()
```

### 2.2 Visualizing GDP vs. Outbound Flights
This scatter plot helps us visualize the trend between a country's wealth and the travel frequency of its residents.

```python
plt.figure(figsize=(12, 6))
sns.regplot(data=df, x='GDP_pc', y='Outbound_Per_Capita', 
            scatter_kws={'alpha':0.5, 'color':'blue'}, 
            line_kws={'color':'red'})
plt.title('GDP per Capita vs. Outbound Flights per Capita (EU 2016-2024)')
plt.xlabel('GDP per Capita (Current USD)')
plt.ylabel('Outbound Flights per Capita')
plt.show()
```

## 3. Hypothesis Testing
We use the **Pearson Correlation Coefficient** to test if the observed relationship is statistically significant.

- **H0 (Null Hypothesis):** There is no correlation between GDP per capita and Outbound flights per capita.
- **H1 (Alternative Hypothesis):** There is a significant positive correlation between GDP per capita and Outbound flights per capita.

```python
# Data Cleaning for Testing
clean_test_df = df.dropna(subset=['GDP_pc', 'Outbound_Per_Capita'])

# Pearson Correlation Test
pearson_coef, p_value = stats.pearsonr(clean_test_df['GDP_pc'], clean_test_df['Outbound_Per_Capita'])

print(f"--- Statistical Test Results ---")
print(f"Pearson Correlation Coefficient: {pearson_coef:.4f}")
print(f"P-value: {p_value:.4e}")

alpha = 0.05
if p_value < alpha:
    print("\nDecision: Reject the Null Hypothesis (H0).")
    print("Evidence suggests a statistically significant correlation.")
else:
    print("\nDecision: Fail to reject the Null Hypothesis (H0).")
```
